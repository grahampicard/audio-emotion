```python
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
```

# Basics
In the case of music, sound waves are generally moving air. We can think of there being some fundamental wave parameters that describe waves.

Let's start off by generating a sin wave.

```python
# four "seconds" of audio
time = np.arange(0, 15, 1 / 22050)
position = np.array([np.sin(440  * 1.5 * np.pi *  x) for x in time])
```

Let's plot our sound wave

```python
fig, ax = plt.subplots(figsize=(20,4))
ax.plot(time, position)
plt.show()
```

So, let's make a guitar tuner

```python
Elow = 82.41
A = 110 
D = 146.83
G = 196
B = 246.94
C = B * (2 ** (1 / 12))
E = 329.63
```

```python
tuner = [Elow, A, D, G, B, C, E]
names = ['Elow', 'A', 'D', 'G', 'B', 'C', 'E']
freqs = {}

for note, name in zip(tuner, names):
    freqs[name] = np.array([np.sin(2 * note * np.pi *  x) for x in time])
```

```python
ipd.Audio(freqs['A'] * 2, rate=22050)
```

```python
ipd.Audio(freqs['D'] + freqs['B'], rate=22050)
```

```python
ipd.Audio((freqs['B'] + (freqs['D']) + freqs['G']), rate=22050)
```

```python
ipd.Audio((freqs['C'] + (freqs['D'] * 4) + freqs['G']), rate=22050)
```
